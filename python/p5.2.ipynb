```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.genfromtxt('stockholm_tmp.dat')
```

april avg from 1800 to now

```python
%matplotlib inline
fig1 = plt.figure()
axes1 = fig1.add_subplot(111)
fig2 = plt.figure()
axes2 = fig2.add_subplot(111)
```

```python
axes1.cla()
axes1.bar(np.arange(1, 13), [np.mean(data[data[:, 1] == i, 3]) for i in range(1, 13)])
axes1.set_title('mean t in different months')
fig1
```

```python
axes1.cla()
axes1.hist(np.arange(1, 13), bins=12, weights=[np.mean(data[data[:, 1] == i, 3]) for i in range(1, 13)])
axes1.set_title('mean t in different months')
fig1
```

```python
april = data[(data[:, 1] == 4) * (data[:, 0] >= 1800), :][:, (0, 2, 3)]
x = np.unique(april[:, 0])
y = map(np.mean, np.split(april[:, 2], np.where(np.diff(april[:,0]))[0] + 1))

axes2.cla()
axes2.plot(x, list(y))
fig2
```
