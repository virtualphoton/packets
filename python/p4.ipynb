```python
import matplotlib.pyplot as pltw
import numpy as np
```

```python
def foo(t):
    return np.arctan(-.0012*t**3+0.4*t**2+.616*t+6120)+.65*np.sin(.24*t+1.23)-.27*np.cos(.21*t-.17)-np.sin(.34*t+.16)/(1+.03*(t-370.5)**5)
    
def plot_foo(t):
    fig, ax = plt.subplots()
    ax.plot(t, foo(t))

def bin_search(t, left, right, eps=.01):
    l0, g0 = sorted((left, right), key=foo)
    for _ in range(100):
        mid = (g0+l0)/2
        f_m = foo(mid)
        if f_m > 0:
            g0 = mid
        elif f_m < 0:
            l0 = mid
        if abs(f_m) < eps:
            break
    return mid, abs(f_m)
```

```python
plot_foo(np.arange(0, 500, .1))
```

```python
plot_foo(np.arange(368.4, 368.5, 1e-8))
```

```python
bin_search(np.arange(0, 1000, .0001), 0, 600, .00000001)
```
